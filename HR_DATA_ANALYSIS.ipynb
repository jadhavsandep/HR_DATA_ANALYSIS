```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('HR_Analytics.csv')
```

```python
data['Department'].unique()
```

```python
data.dropna(inplace=True)
```

```python
data.drop_duplicates(inplace=True)
```

```python
# Histogram
data['Department'].hist()
plt.show()
```

```python
# Boxplot
sns.boxplot(x='EducationField', data=data)
plt.show()
```

```python
# Bar Chart - Attrition Rates by Department
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x="Department", hue="Attrition")
plt.title("Attrition Rates by Department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
#  Pie Chart - Distribution of Attrition (Yes vs. No)
plt.figure(figsize=(6, 6))
attrition_counts = data['Attrition'].value_counts()
plt.pie(attrition_counts, labels=attrition_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Attrition Distribution")
plt.show()
```

```python
#  Histogram - Age Distribution of Employee
plt.figure(figsize=(8, 6))
plt.hist(data['Age'], bins=10, edgecolor='black')
plt.title("Age Distribution of Employees")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# Boxplot - Daily Rate Across Age Groups
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x="AgeGroup", y="DailyRate")
plt.title("Daily Rate Distribution Across Age Groups")
plt.xlabel("Age Group")
plt.ylabel("Daily Rate")
plt.show()
```

```python
# Line Chart - Average Years at Company by Age
avg_years_at_company = data.groupby("Age")["YearsAtCompany"].mean()
plt.figure(figsize=(8, 6))
plt.plot(avg_years_at_company.index, avg_years_at_company.values, marker='o')
plt.title("Average Years at Company by Age")
plt.xlabel("Age")
plt.ylabel("Average Years at Company")
plt.grid()
plt.show()
```

```python
#Scatter Plot - Distance from Home vs. Years at Company
plt.figure(figsize=(8, 6))
plt.scatter(data['DistanceFromHome'], data['YearsAtCompany'], alpha=0.7)
plt.title("Distance from Home vs. Years at Company")
plt.xlabel("Distance from Home")
plt.ylabel("Years at Company")
plt.grid()
plt.show()
```

```python
#Heatmap - Correlations Between Numerical Variables
correlation_matrix = data[["Age", "YearsAtCompany", "DailyRate", "DistanceFromHome"]].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
 # Stacked Bar Chart - Attrition by Business Travel Category
attrition_travel = data.groupby(['BusinessTravel', 'Attrition']).size().unstack()
attrition_travel.plot(kind='bar', stacked=True, figsize=(8, 6))
plt.title("Attrition by Business Travel Category")
plt.xlabel("Business Travel Category")
plt.ylabel("Count")
plt.legend(title="Attrition")
plt.xticks(rotation=45)
plt.show()
```

```python
# Violin Plot - Years at Company by Attrition
plt.figure(figsize=(8, 6))
sns.violinplot(data=data, x="Attrition", y="YearsAtCompany")
plt.title("Years at Company Distribution by Attrition")
plt.xlabel("Attrition")
plt.ylabel("Years at Company")
plt.show()
```

```python
 #Bubble Chart - Daily Rate by Age and Years at Company
plt.figure(figsize=(8, 6))
plt.scatter(data['Age'], data['YearsAtCompany'], s=data['DailyRate'] / 10, alpha=0.5)
plt.title("Bubble Chart: Daily Rate by Age and Years at Company")
plt.xlabel("Age")
plt.ylabel("Years at Company")
plt.grid()
plt.show()
```
